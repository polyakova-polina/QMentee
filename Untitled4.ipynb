Амплитуды содержат информацию о вероятности того, что мы найдем кубит в определенном состоянии, но как только мы измерили кубит, мы с уверенностью знаем, в каком состоянии он находится. Например, если мы измеряем кубит в определенном состоянииесли, то если еще раз измерить, есть 100% шанс найти кубит в этом состоянии. Это означает, что акт измерения изменяет состояние кубитов.


Инициализируем кубит в суперпозиции:

```python
qc = QuantumCircuit(1) # We are redefining qc
initial_state = [0.+1.j/sqrt(2),1/sqrt(2)+0.j]
qc.initialize(initial_state, 0)
```

```python
sim = Aer.get_backend('aer_simulator')  # Tell Qiskit how to simulate our circuit
```

```python
qc.save_statevector()   # Tell simulator to save statevector
qobj = assemble(qc)     # Create a Qobj from the circuit for the simulator to run
result = sim.run(qobj).result() # Do the simulation and return the result
```

```python
out_state = result.get_statevector()
out_state.draw('latex') # Display the output state vector
```

```python
qc.draw('mpl')
```

Здесь видно, что кубит инициализирован в состоянии [0.+0.70710678j 0.70710678+0.j], которое мы и ожидали.

Теперь создадим схему, в которой будем измерять этот кубит:

```python
qc = QuantumCircuit(1) # We are redefining qc
initial_state = [0.+1.j/sqrt(2),1/sqrt(2)+0.j]
qc.initialize(initial_state, 0)
qc.measure_all()
qc.save_statevector()
qc.draw('mpl')
```

Когда мы моделируем всю эту схему, мы видим, что одна из амплитуд всегда равна 0:

```python
qobj = assemble(qc)
state = sim.run(qobj).result().get_statevector()
print("State of Measured Qubit = " + str(state))
```

Можно повторно запустить эту ячейку несколько раз, чтобы повторно инициализировать кубит и измерить его снова. Оба исхода равновероятны, но состояние кубита никогда не является суперпозицией. Несколько интересно, что глобальная фаза состояния сохраняется, но поскольку это глобальная фаза, мы никогда не сможем измерить ее на реальном квантовом компьютере.

Мы видим, что для записи состояния кубита требуется отслеживать два комплексных числа, но при использовании настоящего квантового компьютера мы всегда будем получать ответ «да» или «нет» (0 или 1) для каждого кубита.
При использовании настоящего квантового компьютера мы не можем видеть состояния наших кубитов в середине вычислений, так как это уничтожит их
